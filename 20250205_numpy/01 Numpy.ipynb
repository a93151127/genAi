# Numpy

#### NumPy簡介

NumPy 是 Python 中一個強大的數值計算庫，廣泛應用於數據科學、機器學習以及科學計算領域。它提供了高效的多維陣列物件 `ndarray`，以及大量操作這些陣列的數學函數。

#### 核心功能

1. **多維陣列（ndarray）**  
   NumPy 的核心是其提供的 `ndarray`，它是一種固定大小、相同類型的多維陣列，比 Python 的內建列表（list）更高效，並且支持向量化操作，能夠極大提高數據處理的效率。

2. **數學運算**  
   NumPy 提供了許多數學函數來對陣列進行基本運算，如加法、減法、乘法、除法等。這些運算可以直接應用於整個陣列，而不需要顯式的迴圈。

3. **廣播（Broadcasting）**  
   NumPy 支持廣播機制，可以讓形狀不同的陣列進行運算，這樣在處理數據時不需要顯式地調整陣列大小。

4. **隨機數生成**  
   NumPy 還提供了強大的隨機數生成模組，用於生成各種分布的隨機數，對於模擬和統計分析非常有用。

5. **線性代數功能**  
   NumPy 內建了線性代數運算，如矩陣乘法、特徵值分解、行列式計算等。

6. **性能優化**  
   NumPy 是基於 C 語言編寫的，因此它的運算速度比 Python 本身的數據結構要快得多，並且支持並行計算。

#### NumPy 中的「維度」概念
在 NumPy 中，最基本的數據結構是「n維陣列（ndarray）」，因此理解「維度」是非常重要的。維度代表了數據陣列的結構，它告訴我們陣列的形狀和組織方式。簡單來說，維度是用來描述陣列的空間大小和層次。

維度的意義
* 0 維陣列（純量）：只有一個數值，沒有行列。例如，5 是一個 0 維陣列。

* 1 維陣列（向量）：只有一行或一列的陣列，像是一個數字序列。例如，[1, 2, 3] 是一個 1 維陣列。

* 2 維陣列（矩陣）：有多行多列的陣列，形狀像是矩陣。例如，[[1, 2], [3, 4]] 是一個 2 維陣列。

* 3 維陣列（立體陣列）：由多個矩陣組成的陣列，可以看作是包含多層矩陣的結構。例如，一個 3x2x2 的陣列就包含了 3 個 2x2 的矩陣。

維度不僅僅是表示陣列有幾層，它還告訴我們每層的大小與形狀。

### NumPy 中的 `dtype`

在 NumPy 中，`dtype` 是指數據類型（Data Type）。它用來描述 NumPy 陣列中元素的數據類型，從而決定了這些元素如何存儲和操作。了解 `dtype` 是處理數據時非常重要的，因為它會影響運算效率、內存使用以及支持的數學操作。

#### 常見的 `dtype`

NumPy 支持多種數據類型，常見的包括：

- **整數類型**：
  - `np.int8`：8 位整數
  - `np.int16`：16 位整數
  - `np.int32`：32 位整數
  - `np.int64`：64 位整數
- **浮點數類型**：
  - `np.float16`：16 位浮點數
  - `np.float32`：32 位浮點數
  - `np.float64`：64 位浮點數
- **複數類型**：
  - `np.complex64`：64 位複數
  - `np.complex128`：128 位複數
- **布林類型**：
  - `np.bool_`：布林值（True 或 False）
- **其他類型**：
  - `np.str_`：字符串
  - `np.object_`：Python 物件類型


#### 為什麼 dtype 很重要？
* 性能優化：選擇合適的數據類型可以有效地減少記憶體使用，並加速計算。例如，使用 np.float32 而不是 np.float64 可以減少記憶體佔用，並加快計算。
* 數據一致性：確保所有元素都具有相同的數據類型，有助於避免數據轉換錯誤。
* 支持特定運算：不同的數據類型支持不同的數學操作和精度範圍。選擇合適的數據類型可以避免計算過程中的溢出或精度損失。

````python
# 自動推斷 dtype
arr = np.array([1, 2, 3])
print(arr, arr.dtype)  # 輸出: int64（取決於系統架構）

# 顯式指定 dtype
arr_float = np.array([1, 2, 3], dtype=np.float32)
print(arr_float, arr_float.dtype)  # 輸出: float32
````

## Slice（切片） & Stride（步長）

在 NumPy 中，切片是對陣列的一種操作方式，它允許你選擇陣列的一部分來進行處理。切片使用 `:` 符號，並且可以選擇開始位置、結束位置和步長。

基本語法

### `array[開始: 結束]`
這是最簡單的切片語法，其中：
- **開始** 是切片的起始位置（包括該位置的元素），如果省略，則默認從陣列的開頭開始。
- **結束** 是切片的結束位置（不包括該位置的元素），如果省略，則默認取到陣列的最後一個元素。

### 步長（Stride） `array[開始: 結束: 步長]`
步長（stride）控制了每次切片時跳過多少元素。這意味著在切片過程中，每次會選擇元素後，跳過一定數量的元素。

步長 是切片間隔的大小。如果步長為 2，則每隔一個元素選擇一個元素。如果省略步長，默認為 1。

NOTE: 省略開始位置即為從頭開始，省略結束位置即為取至最後一個

NOTE: 不包含結束index位置的數值

1D array

````python
x = np.arange(6)  # array([0, 1, 2, 3, 4, 5])
x[1:5]   # [1, 2, 3, 4]
x[1:] # x[1:6]: [1, 2, 3, 4, 5]
x[:3] # x[0:3] , [0, 1, 2]
x[1:4:2] # [1, 3], stride=2 每次跳2個位置
````

2D array

````python
x = np.array([
    [0, 1, 2],
    [3, 4, 5]
])

x[0, 0:2] # [0, 1]

x[:, 1:]
# [[1, 2],
#  [4, 5]]

x[::1, ::2]
# [[0, 2],
#  [3, 5]]
````

特殊用法: Reverse axis 某個軸反向排列

````python
# e.g. BGR -> RGB
x = np.array([
    [0, 1, 2],
    [3, 4, 5]
])
x_reverse = x[:, ::-1]

print('original: \n', x)
print('reverse axis1: \n', x_reverse)
# [[2, 1, 0],
#  [5, 4, 3]]
````

Ellipsis(`[...]`): 省略多個`[:, :]`

## 條件篩選 & 賦值: Boolean / Mask Index

這些操作可以使用布林（Boolean）索引或遮罩（Mask）索引來實現。這些操作通過布林運算符來進行篩選、過濾和賦值。

`&`: and

`|`: or

`~`: not

````python
x = np.arange(6)
x[x<4] # 篩選出<4的元素
````

````python
# 也可將條件式寫成變數，因條件式計算完後會得到一個bool mask
condition = x<4
print(condition)
x[condition] # [0, 1, 2, 3]
````

````python
# 串連條件: and, or, not
print('>=1 and <4:', x[(x>=1) & (x<4)])
print('<=1 or >4 :', x[(x<=1) | (x>4)])
print('not > 4   :', x[~(x>4)])
````

````python
# 條件賦值
x = np.arange(6)

x[x<4] = 0

condition = x<4
x[condition] = 0

x
````

## NumPy 的 Broadcasting（廣播機制）

在 NumPy 中，Broadcasting 是一種強大的功能，允許不同形狀的陣列進行數學運算，而不需要顯式地調整它們的形狀。廣播機制通過自動擴展較小的陣列，使其形狀與較大的陣列對齊，從而進行逐元素的運算。

NumPy 通常是基於逐個元素的運算。在最基本的情況下，這兩個陣列的形狀必須完全相同

````python
a = np.array([1.0, 2.0, 3.0])
b = np.array([2.0, 2.0, 2.0])
a * b
````

當陣列與常量值進行運算時，常量會被自動擴展以匹配陣列的形狀

image.png

````python
import numpy as np
a = np.array([1.0, 2.0, 3.0])
b = 2.0
a * b
````

廣播規則

當對兩個陣列進行運算時，NumPy 會逐元素比較它們的形狀。比較從最後一個維度（即最右邊的維度）開始，然後依次向左進行。如果兩個維度是相容的，則當：

1. 它們相等，或者
2. 其中一個維度為 1。

如果這些條件不滿足，則會引發異常，並提示陣列形狀不相容

**例1**

假設有一個包含 RGB 值的3D陣列，並且希望將圖像中的每種顏色按不同的值進行縮放，可以將圖像與一個具有 3 個值的 1 維陣列相乘。根據廣播規則，將這些陣列的尾部軸的大小對齊，顯示它們是相容的
```
Image  (3d array): 256 x 256 x 3
Scale  (1d array):             3
Result (3d array): 256 x 256 x 3
```
**例2**

A 和 B 陣列都有一個長度為 1 的軸，這些軸在廣播操作過程中會被擴展為更大的大小
```
A      (4d array):  8 x 1 x 6 x 1
B      (3d array):      7 x 1 x 5
Result (4d array):  8 x 7 x 6 x 5
```

以下的A B 陣列是可互相廣播的
```
A      (2d array):  5 x 4
B      (1d array):      1
Result (2d array):  5 x 4

A      (2d array):  5 x 4
B      (1d array):      4
Result (2d array):  5 x 4

A      (3d array):  15 x 3 x 5
B      (3d array):  15 x 1 x 5
Result (3d array):  15 x 3 x 5

A      (3d array):  15 x 3 x 5
B      (2d array):       3 x 5
Result (3d array):  15 x 3 x 5

A      (3d array):  15 x 3 x 5
B      (2d array):       3 x 1
Result (3d array):  15 x 3 x 5
```

以下為**不可**廣播的
```
A      (1d array):  3
B      (1d array):  4 # 最右不符合

A      (2d array):      2 x 1
B      (3d array):  8 x 4 x 3 # 右邊數來第2個不符合
```

## 合併或堆疊多個陣列

### np.concatenate: 串接array

沿著某個軸(axis)串接合併，合併之後的陣列維度會跟原本的維度數量相同，不會額外增加新的維度。

注意: 除了要合併的維度之外，其餘的維度大小都要一樣喔!



image.png

image.png

````python
x = np.array([
    [0, 1, 2],
    [3, 4, 5]
])
y = np.array([
    [6, 7, 8]
])
z = np.array([
    [55],
    [66]
])
print('x.shape', x.shape) # (2, 3)
print('y.shape', y.shape) # (1, 3)
print('z.shape', z.shape) # (2, 1)

a = np.concatenate((x, y), axis=0)
print('concat(x, y) shape:', a.shape) # (3, 3)
print(a)
# [[0 1 2]
#  [3 4 5]
#  [6 7 8]]
b = np.concatenate((x, z), axis=1)
print('concat(x, z) shape:', b.shape) # (2, 4)
print(b)
# [[ 0  1  2 55]
#  [ 3  4  5 66]]
````

### stack: 堆疊陣列

與concatenate不同的地方在於，需要合併的陣列無論是維度或是各維度的長度都需要相同，另外合併後的陣列會額外新增一個維度。

而vstack(vertical stacking)與hstack(horizontal stacking)則與concatenate功能類似，只是不需要指定合併的維度

image.png

````python
# 創建三個2D array
a = np.array([[0, 1], [2, 3]])
b = np.array([[4, 5], [6, 7]])
c = np.array([[8,  9], [10, 11]])
a.shape, b.shape, c.shape
````

````python
# 將三個陣列在第0個維度堆疊起來
s = np.stack([a, b, c], axis=0)
print(s)
print(s.shape)
````

````python
# 將三個陣列沿著第0個維度垂直合併起來
v = np.vstack([a, b, c])
print(v)
print(v.shape)
````

````python
# 將三個陣列沿著第1個維度合併起來
h = np.hstack([a, b, c])
print(h)
print(h.shape)
````

### 重複拼貼陣列元素 repeat & tile

np.repeat()

````python
x = np.array([[0, 1], [2, 3]])
x
````

````python
# 重複陣列中的元素兩次，預設會將陣列攤平成一維陣列
np.repeat(x, repeats=2)
# [0, 0, 1, 1, 2, 2, 3, 3]
````

````python
# 指定每個元素的重複次數
np.repeat(x, repeats=(1, 2, 3, 4))
````

````python
# 以axis參數指定要重複時所依據的維度
# 指定axis = 0，numpy就會將第一軸index相同的元素(ex.[0, 1])當作同一組做重複。
np.repeat(x, repeats=2, axis=0)
# [[0, 1],
#  [0, 1],
#  [2, 3],
#  [2, 3]]
````

````python
# 以axis參數指定要重複時所依據的維度
# 指定axis = 1，numpy就會將第二軸index相同的元素當作同一組做重複。
# (ex.[0
#   2])

np.repeat(x, repeats=2, axis=1)
# [[0, 0, 1, 1],
#  [2, 2, 3, 3]]
````

np.tile()

````python
x = np.array([[0, 1], [2, 3]])
print(x)
````

````python
# 預設沿著axis=1增加
np.tile(x, reps=2)
# [[0, 1, 0, 1],
#  [2, 3, 2, 3]]
````

````python
# reps參數可以指定每個維度的重複次數
np.tile(x, reps=(1, 2))
````

````python
# 設定在第0個維度重複兩次，第1個維度不做重複動作
np.tile(x, reps=(2, 1))
````

````python
# 若reps參數中填入的維度大於原本陣列的維度，則會自動增加陣列維度
y = np.tile(x, reps=(2, 2, 2))
print(y.shape)
print(y)
````

## 排序array

針對陣列中的元素，可以進行排序，也可以利用axis參數讓功能更加彈性

````python
# 創建一個二維陣列
x = np.array([[6, 5, 4], [1, 2, 3]])
print(x)
````

````python
# 使用np.sort排序陣列內元素，預設會沿著最後一個維度做排序
np.sort(x)
````

````python
# 也可以直接使用axis參數指定要依照哪個維度做排序
# 在此範例中因為x的維度有(0,1)兩種可能性，指定axis=-1 將等於 axis=1
np.sort(x, axis=-1)
````

````python
# 依據第0個維度做排序
np.sort(x, axis=0)
# [[1, 2, 3],
#  [6, 5, 4]]
````

````python
# 若axis = None，則會攤平並排序陣列中所有元素
np.sort(x, axis=None)
# [1, 2, 3, 4, 5, 6]
````

````python
# 若axis = None，則會攤平並排序陣列中所有元素
np.sort(x, axis=None, )
# [1, 2, 3, 4, 5, 6]
````

np.argsort()

返回陣列排序後的索引位置的函數，而不是直接返回排序後的陣列。這些索引表示排序後元素在原陣列中的位置，因此我們可以使用這些索引來對原始陣列進行重排或進行其他操作。

````python
arr = np.array([3, 1, 4, 1, 5, 9, 2])

# 返回排序後的索引
sorted_indices = np.argsort(arr)

print(sorted_indices)
````

````python
# 使用 argsort 的索引重新排列原始陣列
sorted_arr = arr[sorted_indices]

print(sorted_arr)
# 輸出: [1 1 2 3 4 5 9]
````

````python
# 獲得遞減排序的索引
sorted_indices_desc = np.argsort(arr)[::-1]

# 使用遞減排序的索引
sorted_arr_desc = arr[sorted_indices_desc]

print(sorted_arr_desc)
# 輸出: [9 5 4 3 2 1 1]
````

## 線性代數模組

NumPy 提供了一個強大的 線性代數模組，可以用來處理矩陣運算、解線性方程、計算行列式、特徵值分解等常見的數學問題。這些功能主要集中在 numpy.linalg 模組中，並且支持大部分線性代數運算，對於科學計算、數據分析和機器學習等領域非常有用

````python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
result1 = np.dot(A, B)
result2 = A @ B
print(result1)
print(result2)
````

````python
# 逆矩陣（Inverse Matrix）
A = np.array([[0, 1], [2, 3]])
inv_A = np.linalg.inv(A)
print(inv_A)
````

````python
# 將x與x的反矩陣進行內積，結果應為單位矩陣
print(np.dot(A, inv_A))
print(A @ inv_A)
````
