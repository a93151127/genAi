#### 數值型態

**變數類型與範例**  

- `bool`: **Boolean（布林值）**  
  - `True`  
  - `False`  

- `int`: **Integer（整數）**  
  - `5566`  
  - `10`  

- `float`: **Float（浮點數）**  
  - `3.1415926`  
  - 科學記號：`1.0e3`（表示 `1.0 * 10^3`）  

轉換變數類型

str, int, float

```python
a = int(input("a: ")) # 將使用者輸入轉成整數
b = float(input("b: ")) # 將使用者輸入轉成浮點數

print(a+b)
print(type(a), type(b))
```

bool

```python
print(bool(True))   # 布林值：True
print(bool(1))      # 非零整數為真：True
print(bool(45))     # 非零整數為真：True
print(bool(-45))    # 非零整數為真：True

print(bool(False))  # 布林值：False
print(bool(0))      # 整數 0 為假：False
print(bool(0.0))    # 浮點數 0.0 為假：False
```

int

```python
print(5566)       # 整數：5566
print(+5566)      # 正號（無影響）：5566
print(-5566)      # 負號：-5566

print(1,000,000)  # 逗號視為分隔多個值：(1, 0, 0)

million = 1_000_000  # 下劃線作為數字分隔符
print(million)    # 仍然是 1000000（與逗號不同）

print(5_5_6_6)    # 下劃線可用於提高可讀性，實際數值不變：5566
```

類型轉換

```python
print(int(True))      # 布林值轉整數：True 轉換為 1
print(int(False))     # 布林值轉整數：False 轉換為 0

print(bool(1))        # 非零數值為真：True
print(bool(0))        # 0 為假：False

print(int(99.9))      # 浮點數轉整數（去尾法）：99
print(int(1.0e4))     # 科學記數法轉整數：10000

print(bool(1.0))      # 非零浮點數為真：True
print(bool(0.0))      # 0.0 為假：False

print(int('99'))      # 字串轉整數：99
print(int('-20'))     # 帶負號的字串轉整數：-20
print(int('+23'))     # 帶正號的字串轉整數：23
print(int('1_000_000'))  # 帶下劃線的字串轉整數（Python 允許）：1000000
```

浮點數float

```python
print(6.)       # 浮點數表示法（自動補 0）：6.0
print(6.0)      # 浮點數：6.0
print(06.0)     # Python 3 不允許前導 0，會報錯（在 Python 2 可行）
print(6e0)      # 科學記數法（6 × 10^0）：6.0
print(6.0e1)    # 科學記數法（6.0 × 10^1）：60.0

print(float(True))   # 布林值轉浮點數：True → 1.0
print(float(False))  # 布林值轉浮點數：False → 0.0

print(float(56))     # 整數轉浮點數：56 → 56.0
print(float('56'))   # 字串轉浮點數：'56' → 56.0

print(30 + 2.)  # 整數與浮點數相加，自動轉為浮點數：32.0

print(False + 0)   # False 轉換為 0，相加結果為 0
print(False + 0.)  # False 轉換為 0.0，相加結果為 0.0

print(True + 0)    # True 轉換為 1，相加結果為 1
print(True + 0.)   # True 轉換為 1.0，相加結果為 1.0
```
