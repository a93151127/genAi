# Pandas

**Pandas** 是一個強大的 Python 資料處理套件，專門用來處理表格型資料，並提供高效且靈活的數據操作功能。它主要提供兩個資料結構：`DataFrame` 和 `Series`，用來儲存、操作和分析資料。

1. **`DataFrame`**：這是Pandas中最常用的資料結構，類似於Excel的工作表或SQL資料表。它由row和column組成，支援不同資料類型的混合，如數字、字串等。

2. **`Series`**：這是Pandas中的一維資料結構，類似於Numpy中的`array`，它可以儲存單一資料類型，並且有索引來標識每個元素。

Pandas最吸引人的地方之一是其與`Numpy`的緊密結合，底層實現基於`Numpy`的`array`，因此許多在Numpy中使用的操作和函數都可以無縫應用到Pandas的物件上，使資料操作更加高效和便捷。

Pandas支援各種資料處理操作，包括：
- 資料清洗：缺失值處理、資料格式轉換等
- 資料過濾：依條件選擇資料
- 資料運算：統計分析、數值計算等
- 資料匯入與匯出：支援CSV、Excel、SQL等格式

簡單來說，Pandas是資料科學和分析中不可或缺的工具，能讓你高效地處理和分析資料，並輕鬆完成各種資料處理任務。

#### Pandas `iloc` 和 `loc` 的差異

在 **Pandas** 中，`iloc` 和 `loc` 是用來進行資料選擇和過濾的兩個方法，但它們的使用方式和目標不同。以下是它們的主要差異：

---

1. **`iloc`：基於整數位置（Index Position）**
- 使用 **數值索引**（位置）來選取資料。
- 適合用來根據 **row或column** 來提取資料。
- 索引值從 **0 開始**，不管原始的索引標籤為何。

2. **`loc`：基於標籤（Label）**
- 使用 **標籤**（index 或 column name）來選取資料。
- 適合用來根據row或column的 **名稱** 來進行選擇。
- 包括 **區間的結尾**（標籤索引）。

3. **主要差異點**

| 功能                | `iloc`                            | `loc`                      |
|---------------------|----------------------------------|---------------------------|
| 索引方式             | 整數位置（index position）     | 標籤（label）             |
| 區間選取行為         | 不包括結尾（右開區間 `[start:end)`） | 包括結尾（閉區間 `[start:end]`） |
| 支援條件過濾         | 不支援                          | 支援布林條件過濾                |
| 使用場合             | 知道行列的**數字位置**                | 知道行列的**標籤名稱**              |

| 方法                            | 說明                                |
|---------------------------------|-------------------------------------|
| `df.iloc[0, 0]`                | 選取第一欄第一列的資料              |
| `df.loc[0, 'Age']`              | 選取列索引為0且欄位為年齡的資料    |
| `df['Age']` 或 `df.Age`         | 選取年齡欄位的資料 (pd.Series型態) |
| `df[['Age']]`                   | 選取年齡欄位的資料 (pd.DataFrame型態) |


#### 條件篩選

雖然上述的方法可以快速地幫我們挑選資料，但更多時候我們想要篩選的是符合特定規則的資料，例如在成績紀錄內只篩選出期中考不及格的同學們。遇到這種狀況的時候我們可以使用和numpy很類似的用法。

````
# 選擇"身高超過 160" 的資料
tall_people = df[df['height'] > 160]
tall_people
````

````
# 也可以用.loc選擇"身高超過 160" 的資料
tall_people = df.loc[df['height'] > 160]
tall_people
````

````
# 選擇name是Eric的資料
df_eric = df[df["name"]=="Eric"]
df_eric
````

多條件篩選



*   `&`: and
*   `|`: or
* `~`: not



````
# 選擇 身高超過160 且 體重小於65
subset = df[(df['height'] > 160) & (df['weight'] < 65)]
subset
````

````
# 不同寫法
condition1 = df["height"] > 160
condition2 = df["weight"] < 65
subset = df[condition1 & condition2]
subset
````

````
# 選擇 身高超過160 且 體重"不"小於65
condition1 = df["height"] > 160
condition2 = df["weight"] < 65 # >= 65
subset = df[condition1 & ~condition2]
subset
````

df.isin(): 檢查某row的欄位值是否在列表中

````
# 篩選 name 是 'Eric' 或 'Diana' 的人
subset = df[(df["name"] == "Eric") | (df["name"] == "Diana")]
subset = df[df['name'].isin(['Eric', 'Diana'])]
subset
````

#### `DataFrame.apply`

在Pandas中，apply() 方法是一個強大的工具，允許你對 DataFrame 或 Series 的每一列或每一行進行自定義的函數操作。它可以非常靈活地用於資料處理，尤其當你需要對欄位中的數據進行一些自定義運算時，apply() 會非常有用。

````
# 挑選出height與age欄位並且用apply計算每個欄位的平均
# 還記得axis的用法嗎? axis=0代表我們將第0個軸(row)的數值視為同一群，第1個軸(column)的數值則分開計算
df[['age', 'height']].apply(np.mean, axis=0)
````

````
# 利用自定義函式使用apply計算
def calculate_bmi(row):
    height_in_meters = row['height'] / 100  # 將身高轉換為米
    bmi = row['weight'] / (height_in_meters ** 2)  # 計算BMI
    return bmi

df['BMI'] = df.apply(calculate_bmi, axis=1)  # 計算每行的BMI並將結果放入新的列
df
````

#### groupby

```python
DataFrame.groupby(by=None, axis=0, level=None, as_index=True, sort=True, group_keys=True, observed=False, dropna=True)

* by：用於分組的列名或函數。
* as_index：是否將分組列設為索引（預設為 True）。
* sort：是否按照分組鍵排序（預設為 True）。
```

````
# 建立範例資料
data = {
    'branch': ['North', 'North', 'South', 'South', 'East', 'East'],
    'product': ['Coffee', 'Tea', 'Coffee', 'Tea', 'Coffee', 'Tea'],
    'sales': [120, 80, 150, 60, 100, 50],
    'date': ['2023-01-01', '2023-01-01', '2023-01-01', '2023-01-01', '2023-01-01', '2023-01-01']
}

df = pd.DataFrame(data)
df
````

````
# Q: 每個分店的總銷售額是多少？
# 分析：按照 branch 分組，計算 sales 的總和
# 按照 branch 分組: df.groupby("branch")
# 計算sales總和 ["sales"].sum()
branch_sales = df.groupby('branch')['sales'].sum()
branch_sales
````

````
# Q: 每個商品的平均銷售額是多少？
# 分析：按照 product 分組，計算 sales 的平均值
# 按照 product 分組: df.groupby('product')
# sales 的平均值: ['sales'].mean()
product_avg_sales = df.groupby('product')['sales'].mean()
product_avg_sales
````

````
# Q: 每個分店的總銷售額、平均銷售額和最大銷售額是多少？
# 分析：使用 .agg() 執行多種聚合操作 (aggregate)

branch_stats = df.groupby('branch')['sales'].agg(['sum', 'mean', 'max'])
branch_stats
````

````
# Q: 每個分店中，不同商品的銷售總額是多少？
# 分析：同時對 branch 和 product 分組

branch_product_sales = df.groupby(['branch', 'product'])['sales'].sum()
branch_product_sales
````

````
# branch_product_sales.loc[0:1] index已經不是整數 不能用loc[整數]
````

````
# index不是整數, 字串當作index
branch_product_sales.loc["East"]
````

````
# 還是可用iloc[位置index] 取值
branch_product_sales.iloc[0]
````

````
# Q: 只顯示銷售總額超過 150 的分店
# 分析：計算各分店銷售總額 再過濾

high_sales = df.groupby('branch')['sales'].sum()
high_sales_filtered = high_sales[high_sales > 150]
high_sales_filtered
````

#### pivot & pivot_table

```python
DataFrame.pivot(index=None, columns=None, values=None)
```



*   index：用於新資料框中作為索引的欄位。如果未指定，則使用現有的索引，必須為單個欄名
* columns：用於新DataFrame中作為欄的欄位，必須為單個欄名
* values：用於填充新資料框數值的欄位，必須為單個欄名。如果未指定，將使用其餘的所有欄位，並且結果的欄將具有分層索引。

```python
DataFrame.pivot_table(values=None, index=None, columns=None, aggfunc='mean', fill_value=None)
```

*   index：用於新資料框中作為索引的欄位。如果未指定，則使用現有的索引。
* columns：用於新DataFrame中作為欄的欄位
* values：用於填充新資料框數值的欄位。如果未指定，將使用其餘的所有欄位，並且結果的欄將具有分層索引。
* aggfunc：聚合函數，默認為mean ，可改為mean, sum, count等
* fill_value：填補缺失值的數值。


````
# 建立範例資料
data = {
    'branch': ['North', 'North', 'South', 'South', 'East', 'East'],
    'product': ['Coffee', 'Tea', 'Coffee', 'Tea', 'Coffee', 'Tea'],
    'sales': [120, 80, 150, 60, 100, 50],
    'date': ['2023-01-01', '2023-01-01', '2023-01-01', '2023-01-01', '2023-01-01', '2023-01-01']
}

df = pd.DataFrame(data)
df
````

````
# 將資料轉換為每個分店（branch）中不同商品（product）的銷售額
result = df.pivot(index='branch', columns='product', values='sales')
result
````

````
# pivot_table 計算銷售額
result = df.pivot_table(
    index='branch',
    columns='product',
    values='sales',
    aggfunc='sum')
result
````

````
# pivot_table 按日期計算每個分支的總銷售額
pivot_branch_sales_by_date = df.pivot_table(
    values='sales',
    index='date',
    columns='branch',
    aggfunc='sum',
    fill_value=0
)
pivot_branch_sales_by_date
````

````
# 按產品計算分支數量（即銷售紀錄數）
pivot_branch_count_by_product = df.pivot_table(
    values='sales',
    index='product',
    columns='branch',
    aggfunc='count',
    fill_value=0
)
pivot_branch_count_by_product
````

#### melt: 將資料轉回長格式

````
pivot_result = df.pivot(index='branch', columns='product', values='sales')
# 將資料轉換為每個分店（branch）中不同商品（product）的銷售額
result = pivot_result.reset_index().melt(
    id_vars=['branch'],
    var_name='product',
    value_name='sales')
result
````

### 進階資料處理

#### `merge`: 基於鍵值(key)進行合併。類似 SQL 的 JOIN 操作。

````
df1 = pd.DataFrame({
    'id': [1, 2, 3],
    'name': ['Alice', 'Bob', 'Charlie'],
    'age': [25, 30, 35]
})
df1
````

````
df2 = pd.DataFrame({
    'id': [2, 3, 4],
    'city': ['New York', 'Los Angeles', 'Chicago']
})
df2
````

````
# inner: 內連接(交集), 只保留兩個表中鍵值匹配的部分（交集）
result = pd.merge(df1, df2, on='id', how='inner')
result
````

````
# outer: 外連接(聯集), 保留所有資料，缺失值用 NaN 填補
result = pd.merge(df1, df2, on='id', how='outer')
result
````

````
# left: 左連接(交集), 保留左表的所有鍵，右表無匹配則填充 NaN。
result = pd.merge(df1, df2, on='id', how='left')
result
````

````
# right: 右連接(交集), 保留右表的所有鍵，左表無匹配則填充 NaN。
result = pd.merge(df1, df2, on='id', how='right')
result
````

````
# Cross Join（笛卡爾積）: 返回兩表的所有可能組合
result = pd.merge(df1, df2, how='cross')
result
````

### 時間格式資料處理

`datetime` 類型與 `to_datetime`

* `datetime` 是 Pandas 中表示時間的資料類型。

* 使用 `pd.to_datetime` 將文字或其他格式的資料轉換為時間類型。

````
import pandas as pd

# 將字串轉換為 datetime
data = {'date_str': ['2023-01-01', '2023-01-02', '2023-01-03']}
df = pd.DataFrame(data)
df['date'] = pd.to_datetime(df['date_str'])

print(df.dtypes)  # 查看類型
df
````

提取時間屬性

````
df['year'] = df['date'].dt.year       # 提取年份
df['month'] = df['date'].dt.month     # 提取月份
df['day'] = df['date'].dt.day         # 提取日期
df['weekday'] = df['date'].dt.day_name()  # 提取星期名稱

df
````

````
df.set_index('date', inplace=True)
df
````

````
# 以日期當作index選取
df.loc["2023-01-02"]
````

````
# 以日期當作index選取
df.loc["2023-01-02": "2023-01-03"]
````

````
# 取特定月份2023/1月份資料
df.loc["2023-01"]
````
